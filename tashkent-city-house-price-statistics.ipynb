```python
df = df.drop_duplicates()
df.shape
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df.price)
plt.show()
```

```python
df = df[df['price']<200000]
plt.figure(figsize=(10, 5))
sns.histplot(df.price)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df.rooms)
plt.show()
```

```python
df = df[df['rooms']<6]
plt.figure(figsize=(10, 5))
sns.histplot(df.rooms)
plt.show()
```

```python
# df = df[df['price']<200000]
sns.displot(data=df, x='size', kind='kde', palette='Accent', hue='rooms', height=5, aspect=3)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df[df['rooms']<4], x='price', hue='rooms', multiple='stack', palette='Reds', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x="price", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='minor')
plt.minorticks_on()
plt.show()
```

```python
y = np.array(df.price)
plt.figure(figsize=(20, 6))
plt.subplot(131)
plt.plot(range(len(y)), y, '^')
plt.ylabel('price')
plt.xlabel('index')
```

```python
sns.catplot(x='rooms', y='price', data=df, height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price', hue='district')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x='district', y='price')
plt.xticks(rotation=90)
plt.show()
```
