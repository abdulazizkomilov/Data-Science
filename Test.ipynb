# Step 1

```python
import pandas as pd
import numpy as np
import requests
import re
from bs4 import BeautifulSoup
import datetime
from csv import writer
```

```python
url = "https://uznews.uz/uz"
page = requests.get(url)


soup = BeautifulSoup(page.content, 'html.parser')
lists = soup.find_all('div', class_="flex")

with open('words_p.csv', 'w', encoding="utf8", newline="") as f:
  thewriter = writer(f)
  header = ["p", "time"]
  thewriter.writerow(header)
  for list in lists:
    p = list.find('p', class_="height_2_ellips")
    if p:
      text = p.string
      text.split()
      text_done = re.split(pattern=r"[ ]", string = text)
      access_datetime = datetime.datetime.now()
      info = [text_done, access_datetime]
      thewriter.writerow(info)
      
```

# Step 3

```python
import pandas as pd
import string

df = pd.read_csv("/content/words_p.csv")
df.head()
```

```python
df['p']=df['p'].str.lower()
df.head()
```

```python
punct=string.punctuation
def remove_punct(x):
  return x.translate(str.maketrans("","",punct))
df['p']=df['p'].apply(lambda x: remove_punct(x))
df.head()
```

```python
from collections import Counter
cnt = Counter()

for text in df['p'].values:
  for word in text.split():
    cnt[word]+=1
cnt.most_common(10)
```
