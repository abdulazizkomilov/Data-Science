# `**Split dataset into train & test**`

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=44, shuffle =True)
```

```python
#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

# `Defining function For ML models`

```python
def predict(ml_model):          
    model=ml_model.fit(X_train,y_train)
    print('Training score : {}'.format(model.score(X_train,y_train)))
    y_prediction=model.predict(X_test)
    print('predictions are: \n {}'.format(y_prediction))
    print('\n')
    r2_score=metrics.r2_score(y_test,y_prediction)
    print('r2 score: {}'.format(r2_score))
    print('MAE:',metrics.mean_absolute_error(y_test,y_prediction))
    print('MSE:',metrics.mean_squared_error(y_test,y_prediction))
    print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,y_prediction)))
    sns.distplot(y_test-y_prediction)
```

# `Random Forest Algorithm`

```python
predict(RandomForestRegressor(ccp_alpha=0.2,n_estimators=1000))
```

# `Linear Regression Algorithm`

```python
predict(LinearRegression())
```

# `K-Neighbors Algorithm`

```python
predict(KNeighborsRegressor(n_neighbors=3,
    weights='distance',
    algorithm='ball_tree',
    leaf_size=40,
    p=4,
    metric='minkowski',
    metric_params=None,
    n_jobs=None,))
```

# `Test preparation for prediction`

```python
df_test['class'] = df_test['class'].replace({'Economy':0, 'Business': 1})
num_features=[col for col in df_test.columns if df_test[col].dtype!='O']
cat_features=[col for col in df_test.columns if df_test[col].dtype=='O']
categorical=df_test[cat_features]
airline=pd.get_dummies(categorical['airline'], drop_first=True)
source_city=pd.get_dummies(categorical['source_city'], drop_first=True)
destination_city=pd.get_dummies(categorical['destination_city'], drop_first=True)
arrival_time=pd.get_dummies(categorical['arrival_time'], drop_first=True)
df_test = pd.concat([categorical,airline,source_city,destination_city,arrival_time,df_test[num_features]],axis=1)
df_test['stops'] = df_test['stops'].replace({'zero':0, 'one': 1, 'two_or_more':3})
pd.set_option('display.max_columns',35)
df_test.drop(['airline','flight','source_city','departure_time','arrival_time','destination_city'], axis=1, inplace=True)
df_test.head()
```

```python
final_predictions = []
def predict(ml_model):          
    model=ml_model.fit(X_train,y_train)
    print('Training score : {}'.format(model.score(X_train,y_train)))
    y_prediction=model.predict(df_test)
    print('predictions are: \n {}'.format(y_prediction))
    final_predictions.append(y_prediction)
```
